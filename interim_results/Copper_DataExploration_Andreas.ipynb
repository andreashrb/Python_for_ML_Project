```python
import pandas as pd
```

# Einlesen der Daten

```python
# URLs der Daten
url_copper = "https://raw.githubusercontent.com/andreashrb/Python_for_ML_Project/main/data/dataset_copper.csv"
url_economic = "https://raw.githubusercontent.com/andreashrb/Python_for_ML_Project/main/data/copper_economic.csv"
```

## Zeitreihendaten Kupferpreise

```python
df_copper = pd.read_csv(url_copper, index_col=0, decimal=".", thousands=",", parse_dates=['date'])
```

```python
df_copper.head()
```

## Ökonomische Indikatoren für den Kupferpreis

```python
df_economic = pd.read_csv(url_economic, index_col=0, encoding='utf-8')
```

```python
df_economic.head()
```

# Joinen der Datensätze

```python
df_copper.dtypes
```

```python
# Extrahieren des Jahres aus der Datetime-Spalte
df_copper['year'] = pd.DatetimeIndex(df_copper['date']).year
```

```python
df_copper.head()
```

```python
df_copper.dtypes
```

```python
df_economic.dtypes
```

```python
# Joinen der beiden Datensätze per Right-Join
df_copper_economic = pd.merge(left=df_copper, right=df_economic, how='right', left_on='year', right_on='Jahr')
df_copper_economic = df_copper_economic.drop('Jahr', axis = 1)
```

```python
df_copper_economic.head()
```

```python
df_copper_economic.shape
```

Der Datensatz hat 2274 Zeilen (Tagesbasierte Kupferschlusskurse + LME-Bestände). Dies entspricht einer Zeitreine von 2012 - 2020 (8 Jahre).
Insgesamt verfügt der Datensatz über 30 Spalten.

Hinweis: Die Spalten "lme_copper_cash", "lme_copper_three_month", "lme_copper_stock" liegen tagesbasiert vor. Alle anderen Features sind lediglich Jahreswerte zum BIP, der globalen Kupfernachfrage, Produktionsmengen verschiedener Länder, Verbrauchsmengen verschiedener Länder, jahresbezogene Daten zur Nachfrage nach Kupfer und ebenso jährliche Daten zu Kupferreserven verschiedener Länder.

Ursprung der Daten:
1. https://www.westmetall.com/en/markdaten.php?action=table&field=LME_Cu_cash
2. https://de.statista.com/statistik/suche/?q=Kupfer&qKat=search&newSearch=true

### Explorative Zeitreihenanalysen - Tages-Daten

```python
# Zeitreihe der täglichen Kupferpreise

plot = sns.lineplot(x=df_copper_economic['date'],y=df_copper_economic['lme_copper_cash'])
plot.axes.set_title("Entwicklung des Kupferpreises im Zeitverlauf",fontsize=25)
```

**Erkenntnisse:**

Stetiger Fall des Kupferpreises von 2012 bis 2016. Tiefpunkt: Anfang 2016 (<5.000 USD).
Anschließend wieder Anstieg mit starken Einbruch Anfang 2020 (Corona Pandemie), bevor ein sehr starker Anstieg in einer kurzen Zeit festzustellen ist.

Allgemein unterliegt der Kupferpreis starken Schwankungen in teilweise sehr kurzen Zeitzyklen.

```python
# Zeitreihe der monatlichen, durchschnittlichen Kupferpreise

df_copper_economic_month = df_copper_economic_ind.resample('M').mean()

plot_month = sns.lineplot(x=df_copper_economic_month.index,y=df_copper_economic_month['lme_copper_cash'])
plot_month.axes.set_title("Entwicklung des durchschnittlichen, monatlichen Kupferpreises im Zeitverlauf",fontsize=25)
```

**Erkenntnisse:**

Leicht zeitverzögerter Zusammenhang zwischen Anstieg und Fall des Kupferpreises und der Entwicklung der Kupferreserven (gegenläufig).

Die Gegenüberstellung zeigt jedoch, dass die Kupferreserven alleine nicht für die Preisbildung verantwortlich sind. Bsp.: Mitte 2014 liegen die Kupferreserven auf einem sehr niedrigen Niveau, dennoch steigen die Kupferpreise nicht exponentiell.

### Explorative Zeitreihenanalysen - Quartal und Monatlich

```python
# Extrahieren von Jahr, Quartal, Monat und Wochentag aus Datum

df_copper_economic['year'] = df_copper_economic['date'].dt.year
df_copper_economic['quarter'] = df_copper_economic['date'].dt.quarter
df_copper_economic['month'] = df_copper_economic['date'].dt.month
df_copper_economic['weekday'] = df_copper_economic['date'].dt.weekday
```

```python
# Aggregation der Datasets

df_copper_economic_year = df_copper_economic.groupby('year').mean().reset_index()
df_copper_economic_quarter = df_copper_economic.groupby('quarter').mean().reset_index()
df_copper_economic_month = df_copper_economic.groupby('month').mean().reset_index()
df_copper_economic_weekday = df_copper_economic.groupby('weekday').mean().reset_index()
```

```python
# Darstellung der Plots für "Kupferpreis"

fig,(ax1, ax2, ax3)= plt.subplots(nrows=3)
fig.set_size_inches(30, 15)

sns.pointplot(data=df_copper_economic_quarter, x='quarter', y='lme_copper_cash', ax=ax1)
sns.pointplot(data=df_copper_economic_month, x='month', y='lme_copper_cash', ax=ax2)
sns.pointplot(data=df_copper_economic_weekday, x='weekday', y='lme_copper_cash', ax=ax3)

ax1.set_title('Durchschnittlicher Kupferpreis pro Quartal von 2012 - 2020')
ax2.set_title('Durchschnittlicher Kupferpreis pro Monat von 2012 - 2020')
ax3.set_title('Durchschnittlicher Kupferpreis pro Tag der Woche von 2012 - 2020')
```

**Erkenntnisse:**

Quartalsweise: Erstes und letztes Quartal mit deutlich höheren Kupferpreisen als Q2 und Q3

Monatlich: Abgeleitet aus der quartalsweisen Betrachtung weisen die ersten und letzten drei Monate (Januar, Februar, März & Oktober, November, Dezember) die größten Kupferpreise bei einer Betrachtung unabhängig des Jahres auf. Tendenziell am günstigsten scheint der Juni zu sein.

Wochentag: Hier gilt zu beachten, dass die Börse nur von Montag bis Freitag geöffnet hat, sodass auch nur diese 5-Tage erfasst werden. Allgemein sind die Preisunterschiede an den Wochentagen über die gesamte Zeitreihe nicht sehr groß (Spanne von ca. 10 US Dollar Differenz).

--> Zusammenfassend: mögliche Saisonalität Sommer vs. Winter könnte zumindest aus der quartalsweisen sowie monatlichen Betrachtung abgeleitet werden.

```python
# Darstellung der Plots für "Kupferreserven"

fig,(ax1, ax2, ax3)= plt.subplots(nrows=3)
fig.set_size_inches(30, 15)

sns.pointplot(data=df_copper_economic_quarter, x='quarter', y='lme_copper_stock', ax=ax1)
sns.pointplot(data=df_copper_economic_month, x='month', y='lme_copper_stock', ax=ax2)
sns.pointplot(data=df_copper_economic_weekday, x='weekday', y='lme_copper_stock', ax=ax3)

ax1.set_title('Durchschnittliche Kupferreserven pro Quartal von 2012 - 2020')
ax2.set_title('Durchschnittliche Kupferreserven pro Monat von 2012 - 2020')
ax3.set_title('Durchschnittliche Kupferreserven pro Tag der Woche von 2012 - 2020')
```

**Erkenntnisse:**

Konträr zu den Kupferpreisen verhalten sich die Kupferreserven:

Sind diese in den ersten und letzten Monaten eines Jahres im Schnitt auf einem sehr niedrigen Niveau von teilweise unter 240.000 Tonnen, so werden über die Sommermonate die Reserven wieder aufgefüllt.
Damit lassen sich ggf. auch die deutlich höheren Preise in den Wintermonaten erklären.

Bei den Wochentagen fällt nur der Montag auf: dieser Tag scheint in der gesamten Zeitreihe mit der geringsten Menge an Kupferreserven behaftet zu sein. Jedoch fallen die Unterschiede nicht sehr stark ins Gewicht (<2.250 Tonnen).

### Produktionsmengen je Land

### Globale Nachfrage vs. Kupferpreis

```python
# Globale jährliche Nachfrage im Vergleich zum Kupferpreis

ax = df_copper_economic_year.plot(x='year', y='Global_Nachfrage in Mio. Tonnen',title='Globale, durchschnittliche jährliche Kupfernachfrage im Vergleich zum durchschnittlichen Kupferpreis')
ax1 = ax.twinx()

df_copper_economic_year.plot(x='year',y='lme_copper_cash',color='black',ax=ax1)
```

**Erkenntnis:**

Die kontinuierlich steigende Nachfrage von 2012 bis 2020 scheint sich zunächst nicht auf den Kupferpreis auszuwirken. So fällt z.B. der Preis trotz einem deutlichen Nachfrageschub von ca. 1,5 Mio. Tonnen von 2013 auf 2014 weiter.

### Globale Nachfrage, Produktions- und Verbrauchsmenge vs. Kupferpreis

```python
# Kalkulation der Produktions- und Verbrauchsmengen gesamt

df_copper_economic_year['production_global'] = df_copper_economic_year[production_columns].sum(axis = 1)
df_copper_economic_year['consumption_global'] = df_copper_economic_year[consumption_columns].sum(axis = 1)

df_copper_economic_year.head(20)
```

```python
# Umrechnung von Spalte "Global_Nachfrage"

df_copper_economic_year['Global_Nachfrage in Mio. Tonnen'] = df_copper_economic_year['Global_Nachfrage in Mio. Tonnen']*1000
df_copper_economic_year.rename(columns = {'Global_Nachfrage in Mio. Tonnen':'demand_global'}, inplace = True)
```

```python
# Globale Nachfrage, Produktions- und Verbrauchsmenge vs. Kupferpreis

ax = df_copper_economic_year.plot(x='year', y='demand_global',title='Globale Nachfrage, Produktions- und Verbrauchsmenge vs. Kupferpreis')
ax1 = ax.twinx()

df_copper_economic_year.plot(x='year',y='production_global',ax=ax)
df_copper_economic_year.plot(x='year',y='consumption_global',ax=ax)
df_copper_economic_year.plot(x='year',y='lme_copper_cash',color='black',ax=ax1)

ax.set_ylabel('Mengen in 1.000 Tonnen')
```

**Erkenntnisse:**

Zu beachten gilt: Die globale Nachfrage ist deutlich höher, da sie die tatsächliche globale Nachfrage darstellt. Globale Produktions- und Verbrauchsmengen sind nur von den oben genannten Ländern enthalten und ergeben in diesem Fall in Summe die globalen Werte.

Bis 2016 steigt die Produktionsmenge sowie der Verbrauch in ungefähr gleichem Maße an. Die Nachfrage bleibt stetig hoch und erfährt kaum Rückgang. Die Preise fallen bis 2016 stark, was auch mit der großen Differenz zwischen Nachfrage und Angebot zusammenhängt.
Auffallend ist, dass anscheinend ein kleiner Produktionsrückgang (=Reduzierung des Gaps Nachfrage-Angebot) bei weiter steigender Nachfrage und Verbrauch, direkt zu einer deutlichen Preissteigerung führen. Die Produktionssteigerung in Folge wirkt sich erst ein Jahr später (2018) auf zurückgehende Preise aus.

Insgesamt scheint es daher vor allem zeitverzögerte Zusammenhänge zwischen den Produktions- und Verbrauchsmengen und dem Verlauf des Kupferpreises zu geben. Die globale Nachfrage wird durch die "globale Verbrauchsmenge" vergleichsweise gut abgebildet, obwohl es sich nur um die 9 größten Konsumnten von Kupfer handelt.
