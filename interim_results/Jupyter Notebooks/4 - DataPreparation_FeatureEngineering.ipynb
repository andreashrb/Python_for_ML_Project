# Einlesen des Datensatzes

```python
df = pd.read_csv('https://raw.githubusercontent.com/andreashrb/Python_for_ML_Project/main/data/copper_economic_enriched_week.csv', sep=',', encoding='utf-8', index_col=0, decimal=".", thousands=",", parse_dates=['date'])
df = df.reset_index()
```

# Feature Extraction mit tsfresh

```python
# !pip install tsfresh
import tsfresh
from tsfresh import extract_features
from tsfresh import select_features
from tsfresh.utilities.dataframe_functions import impute
from tsfresh.utilities.dataframe_functions import roll_time_series
```

## Erstellung eines rollierenden Zeitreihendatensatzes als DataFrame

```python
df_rolled = roll_time_series(df, column_id="id", column_sort="date", max_timeshift=10)
```

```python
df_rolled.head(50)
```

```python
df_rolled.shape
```

Der rollienrende Zeitreihendatensatz hat nun 110.685 Zeilen und weiterhin 32 Spalten. Aus 470 Zeilen wurden, durch den rollierenden Aufbau des Zeitreihendatensatzes, nun 110.685 Zeilen generiert. Sehr umfangreiche Zeitreihen werden auf diesem Weg sehr groß und könnten Probleme bei der Feature Extraction auf einer einfachen lokalen Maschine verursachen.

## Extraktion möglicher Features aus dem rollierenden Zeitreihendatensatz
Aus dem rollierend aufgebauten Zeitreihendatensatz (df_rolled) werden im folgenden alle möglichen Zeitreihenfeatures, die die Library tsfresh unterstützt, generiert.

```python
# Extracting features from time series data
df_features = extract_features(df_rolled, column_id="id", column_sort="date", column_value='lme_copper_cash')
```

```python
df_features2 = extract_features(df_rolled, column_id="id", column_sort="date", column_value='lme_copper_stock')
```

```python
df_features.head()
```

```python
df_features2.head()
```

```python
df_features.shape
```

```python
df_features2.shape
```

Es wurden 798 Features aus den vorliegenden Zeitreihen (Kupferpreis und Kupferreserven) extrahiert!

```python
# Joinen der beiden Datensätze

df_features_final = pd.merge(df_features, df_features2, left_index=True, right_index=True)
```

```python
df_features_final.head()
```

Insgesamt liegen nun 1578 Features zu den beiden Zeitreihen (Kupferpreis und Kupferreserven) vor.

## Selektion relevanter Features mit Bezug zur Zielvariable

```python
# Harmonisierung der Indizierung des ursprünglichen DataFrames und Speicherung der Zielvariable "y"
df = df.set_index(['id', 'date'])
y = df['lme_copper_cash']
y
```

```python
# Aus den vorliegenden Features (789) werden jetzt die für die Vorhersage der Zielvariable potentiell relevanten Features herausgefiltert.

impute(df_features_final)
features_filtered = select_features(df_features_final, y)
```

```python
features_filtered.head()
```

```python
features_filtered.shape
```

Aus den 1578 Features, die initial mit der tsfresh Library für die beiden Zeitreihen erzeugt und berechnet wurden, sind nach der Feature Selection noch 876 Features übrig geblieben.

Analog zum ursprünglichen DataFrame liegen wieder 470 Zeilen im Datensatz vor.

Lt. den Entwicklern der tsfresh-Library kann der DataFrame "features_filtered" nun verwendet werden, um ein Modell zu trainieren.

# Extraktion weiterer Features aus der Zeitreihe

Im Rahmen der Data Exploration wurde analysiert, inwiefern die Quartale, Monate, Kalenderwochen und Wochentage Einfluss auf den Kupferpreis haben. Letztlich ging es dabei primär um die Identifikation saisonaler Effekte, die sich auf den Kupferpreis und dessen Entwicklung auswirken. Gewisse saisonale Effekte sind, bei Betrachtung der gesamten Zeitreihe und der jeweiligen Durchschnittskupferpreise pro Quartal, Monat, Woche und Wochentag, nicht von der Hand zu weisen, auch wenn diese saisonalen Effekte, wie in der DataExploration bereits ausführlicher erläutert, zunächst einmal nicht besonders stark ausgefallen sind. Aus der Autokorrelationsanalyse heraus war sogar von fehlender statistischer Signifikanz die Rede. Dennoch sollen, aufgrund der leichten saisonalen Muster, die explorativ identifiziert werden konnten, auch diese Features wieder Berücksichtigung finden.

```python
# Hinzufügen der Spalten "quarter", "month", "weekofyear" zum DataFrame "features_filtered"

df= df.reset_index()
df['quarter'] = pd.DatetimeIndex(df['date']).quarter
df['month'] = pd.DatetimeIndex(df['date']).month
df['weekofyear'] = pd.DatetimeIndex(df['date']).weekofyear
df['year'] = pd.DatetimeIndex(df['date']).year

# Setzen des Index, analog zum Dataframe "features_filtered"
df = df.set_index(['id','date'])
```

```python
# Abspeichern der relevanten, zu joinenden, Features in separater Variable
df_to_join = df[['year','quarter','month','weekofyear']]
```

```python
# Ausgabe des zu joinenden DataFrames
df_to_join.head()
```

```python
# Für den Join auf den Indexspalten müssen diese identisch benannt sein
features_filtered = features_filtered.rename_axis(['id','date'])
```

```python
# Join
features_filtered = pd.merge(features_filtered, df_to_join, left_index=True, right_index=True)
```

```python
# Ausgabe des gejointen DataFrames
features_filtered.head()
```

```python
from tsfresh.feature_selection.relevance import calculate_relevance_table

relevance_table = calculate_relevance_table(features_filtered, y)
relevance_table = relevance_table[relevance_table.relevant]
relevance_table.sort_values('p_value', inplace = True)

relevance_table.head(877)
```

```python
df_selected = pd.DataFrame()

for i in relevance_table['feature']:
    df_selected[i] = features_filtered[i]
    print(i)
```
